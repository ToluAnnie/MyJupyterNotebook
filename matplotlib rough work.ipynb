```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2)
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (20 * np.random.rand(N))**2
```

```python
plt.subplots(figsize=(14, 6))

plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Spectral')
plt.colorbar()

plt.show()
```

```python
fig, axes = plt.subplots(1,2,figsize=(14, 6))
axes[0].scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel1')



axes[1].scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel2')


plt.show()
```
